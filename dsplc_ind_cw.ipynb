# **Feature Engineering**

```python
df["date"] = pd.to_datetime(df["date"])
```

```python
# 1. Vaccination Coverage Level (based on % of population vaccinated)
df["coverage_level"] = pd.cut(
    df["people_vaccinated_per_hundred"],
    bins=[-1, 25, 50, 75, 100],
    labels=["Low", "Moderate", "High", "Very High"]
)

# 2. Daily Vaccination Intensity (based on vaccinations per million)
df["intensity_level"] = pd.qcut(
    df["daily_vaccinations_per_million"],
    q=4,
    labels=["Slow", "Moderate", "Fast", "Very Fast"]
)

# 3. Vaccination Progress Stage (based on days since first vaccination per loca)
first_date = df["date"].min()
df["days_since_first_vax"] = (df["date"] - first_date).dt.days

df["progress_stage"] = pd.cut(
    df["days_since_first_vax"],
    bins=[-1, 30, 90, 180, 365, float("inf")],
    labels=["Early", "Developing", "Established", "Mature", "Long-term"]
)
```
